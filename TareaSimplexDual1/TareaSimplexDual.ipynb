# Simplex Dual

```python
import pandas as pd
```

```python
original = pd.DataFrame([{'z':1, 'A':-160, 'B':-120, 'C':-80, 's1':0,
                          's2':0, 's3':0, 's4':0, 's5':0, 'RHS': 0},
                         {'z':0, 'A':-1, 'B':-1, 'C':-1, 's1':1,
                          's2':0,'s3':0, 's4':0, 's5':0, 'RHS': -1000},
                         {'z':0, 'A':-1, 'B':0, 'C':0, 's1':0,
                          's2':1, 's3':0, 's4':0, 's5':0, 'RHS': -150},
                         {'z':0, 'A':0, 'B':-1, 'C':0, 's1':0,
                          's2':0, 's3':1, 's4':0, 's5':0, 'RHS': -100},
                         {'z':0, 'A':0, 'B':0, 'C':1, 's1':0,
                          's2':0, 's3':0, 's4':1, 's5':0, 'RHS': 400},
                         {'z':0, 'A':1, 'B':1, 'C':1, 's1':0, 
                          's2':0, 's3':0, 's4':0, 's5':1, 'RHS': 1000}],
                         index=['z', 's1', 's2', 's3', 's4', 's5'],
                         columns=['z','A','B','C',
                                  's1','s2','s3','s4','s5',
                                  'RHS'])
# Show
original
```

```python
# Copiamos para trabajar
wip = original.copy()

# Primera iteración
wip.iloc[1] *= -1
wip.iloc[0] += wip.iloc[1] * 80
wip.iloc[4] += wip.iloc[1] * -1
wip.iloc[5] += wip.iloc[1] * -1

# Cambia variables del pivote
a = list(wip.columns)
a[3] = 's1'
wip.columns = a
a = list(wip.index)
a[1] = 'C'
wip.index = a
# Show
wip
```

```python
# Segunda iterción
wip.iloc[4] *= -1
wip.iloc[0] += wip.iloc[4] * 40
wip.iloc[1] += wip.iloc[4] * -1
wip.iloc[3] += wip.iloc[4]

# Cambia variables del pivote
a = list(wip.columns)
a[2] = 's4'
wip.columns = a
a = list(wip.index)
a[4] = 'B'
wip.index = a

# Show
wip
```

```python
# Tercera iterción
wip.iloc[2] *= -1
wip.iloc[0] += wip.iloc[2] * 40
wip.iloc[3] += wip.iloc[2] * -1
wip.iloc[4] += wip.iloc[2] * -1

# Cambia variables del pivote
a = list(wip.columns)
a[1] = 's4'
wip.columns = a
a = list(wip.index)
a[2] = 'A'
wip.index = a

# Show
wip
```
